```python
# Histogram
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel('../Data/2021girlsnames.xlsx', sheet_name='6', skiprows=6)
```

```python
print(df['Count'].max())
print(df['Count'].mean())
print(df['Count'].median())
print(df['Count'].mode())
```

```python
count = df['Count']
```

```python
sns.set_style("darkgrid")

sns.set(rc={'figure.figsize':(10, 5)})

# Creating variables to align mean and max value annotations
text_offset = 2
text_y_coord = 2900
alpha_hist = 0.6
alpha = 1
text_color = '#61606a'

# Plotting the histogram and kde
sns.histplot(data=count, bins=20, log_scale=(True, False), color='#D24165', alpha=alpha_hist)
plt.xlabel('Number of babies given each name')
    

# Plotting the line and annotations showing the mean
mean = count.mean()
plt.axvline(mean, color=text_color, linestyle='--', label='Mean', alpha=alpha)
plt.text(mean-text_offset, text_y_coord, f'Mean: ({round(mean, 1)})',rotation=90, color=text_color,alpha=alpha, fontsize=10)

# Plotting the line and annotations showing the maximum value
max_char = count.max()
plt.axvline(max_char, color=text_color, linestyle='--', label='Max',alpha=alpha)
plt.text(max_char-(text_offset*100), text_y_coord, f'Max: ({round(max_char, 1)})', rotation=90, color=text_color, alpha=alpha, fontsize=10)

# Plotting the line and annotations showing the mode
mode = df['Count'].mode().item()
plt.axvline(mode, color=text_color, linestyle='--', alpha=alpha)
plt.text(mode-(text_offset/15), text_y_coord, f'Min, Mode: ({round(mode, 1)})',rotation=90, color=text_color,alpha=alpha, fontsize=10)

plt.xticks([10, 100, 1000])
#plt.ylim(0, 3500)
plt.show()
```
