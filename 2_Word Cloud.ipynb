```python
import pandas as pd
from PIL import Image
import numpy as np
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
```

# Importing Data

```python
def import_data(path):
    df = pd.read_excel(path, sheet_name='1', skiprows=6)
    df['inverted_rank'] = 101-df['Rank']
    frequency_dict = df.set_index('Name').to_dict()['inverted_rank']
    return frequency_dict
```

# Wordcloud

```python
from os import path
def make_wordcloud(frequency_dict, shape_path):
    mask = np.array(Image.open(shape_path))   
    wc = WordCloud(background_color="white", mask=mask)
    wc.generate_from_frequencies(frequency_dict)
    
    colouring = np.array(Image.open(shape_path))
    image_colors = ImageColorGenerator(colouring)
    
    plt.figure(figsize=(20,10))
    plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
    plt.axis('off')
    plt.show()
    

#background_color="white"
#background_color=(255,224, 249)
```

# Girls Names

```python
girls_data_path = '../Data/2021girlsnames.xlsx'
girls_shape_path = '../Images/pink_footprint.jpg'


frequency_dict = import_data(girls_data_path)
make_wordcloud(frequency_dict, girls_shape_path)
```

# Boys Names

```python
boys_data_path = '../Data/2021boysnames.xlsx'
boys_shape_path = '../Images/blue_footprint.jpg'


frequency_dict = import_data(boys_data_path)
make_wordcloud(frequency_dict, boys_shape_path)
```
